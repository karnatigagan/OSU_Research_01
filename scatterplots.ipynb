```python
#Import Libraries
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# Extract data and targets 
X = iris.data.features 
y = iris.data.targets 

# Combine features and targets into a single dataframe
iris_df = pd.concat([X, y], axis=1)
  
# Display metadata variable information 
print(iris.metadata) 
print(iris.variables) 

# Display the first few rows of the dataset
print(iris_df.head())

# Create Scatterplots using seaborn
#Sepal length vs. Sepal width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length', y='sepal width', hue='class', data=iris_df, palette='viridis')
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
#Petal Length vs. Petal width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal length', y='petal width', hue='class', data=iris_df, palette='viridis')
plt.title('Scatterplot of Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```
